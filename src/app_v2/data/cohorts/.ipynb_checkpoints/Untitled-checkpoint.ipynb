```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns', 1000)
```

```python
df = pd.read_csv("../texasDashboardData.csv")
df
```

```python
df.columns
```

```python
# Columns of interest
conditions = ['has_diabetes', 'has_heart_failure', 'has_hypertension']
outcomes = ['ever_deceased', 'ever_readmitted']
numeric_cols = conditions + outcomes + ['Age', 'BMI']

# Aggregate at patient-assessment level (mean for numeric/binary flags)
agg_df = df.groupby(['Beneficiary_ID', 'Assessment_Effective_Date'])[numeric_cols].mean().reset_index()
def condition_eda(df, condition):
    print(f"--- EDA for {condition} ---")
    
    # Filter patients with the condition
    condition_df = df[df[condition] > 0]  # >0 in case mean from aggregation
    
    n_patients = len(condition_df)
    print("Number of patient-assessments:", n_patients)
    
    # Outcome summaries
    for outcome in outcomes:
        outcome_counts = condition_df[outcome].value_counts()
        outcome_percent = condition_df[outcome].value_counts(normalize=True) * 100
        print(f"\nOutcome: {outcome}")
        print(pd.DataFrame({'count': outcome_counts, 'percent': outcome_percent.round(2)}))
    
    # Age and BMI stats
    print("\nAge stats:\n", condition_df['Age'].describe())
    print("BMI stats:\n", condition_df['BMI'].describe())

# Run for each condition
for cond in conditions:
    condition_eda(agg_df, cond)
```

```python
for cond in conditions:
    cond_df = agg_df[agg_df[cond] > 0]
    for outcome in outcomes:
        sns.countplot(data=cond_df, x=outcome)
        plt.title(f"{outcome} distribution for {cond}")
        plt.show()
for cond in conditions:
    cond_df = agg_df[agg_df[cond] > 0]
    sns.histplot(cond_df['Age'], bins=20, kde=True)
    plt.title(f"Age distribution for {cond}")
    plt.show()
    
    sns.histplot(cond_df['BMI'], bins=20, kde=True)
    plt.title(f"BMI distribution for {cond}")
    plt.show()
```

```python
summary = []
for cond in conditions:
    cond_df = agg_df[agg_df[cond] > 0]
    summary.append({
        'Condition': cond,
        'N': len(cond_df),
        'Deceased_%': cond_df['is_deceased'].mean() * 100,
        'Readmission_%': cond_df['READMISSION'].mean() * 100,
        'Age_mean': cond_df['Age'].mean(),
        'BMI_mean': cond_df['BMI'].mean()
    })

summary_df = pd.DataFrame(summary)
summary_df
```

```python
full_cohort = df.copy()
diabetes_cohort = df[df['is_diabetes'] == 1]
hf_cohort = df[df['is_heart_failure'] == 1]
htn_cohort = df[df['is_hypertension'] == 1]
```

```python
diabetes_cohort.to_csv('../cohorts/diabetes_v1.csv', index=False)
hf_cohort.to_csv('../cohorts/heartfailure_v1.csv', index=False)
htn_cohort.to_csv('../cohorts/hypertension_v1.csv', index=False)
```
