```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import category_encoders as ce
import seaborn as sns
pd.set_option('future.no_silent_downcasting', True)
```

```python
df_data = pd.read_csv('data/dashboardData_v2.csv', low_memory=False)
df_scores = pd.read_csv('data/patientRiskScores.csv').drop('Age', axis=1)
```

```python
df_final = df_data.merge(
    df_scores,
    left_on='Beneficiary_ID',
    right_on='BENE_ID',
    how='inner'
).drop('BENE_ID', axis=1)
df_final
```

```python
race_cols = [
    "White",
    "Black_or_African_American",
    "Asian",
    "Hispanic_or_Latino",
    "American_Indian_or_Alaska_Native",
    "Native_Hawiian_or_Pacific_Islander"
]
hipps_col = 'Submitted_HIPPS_Code'
icd_cols = [
    'Primary_Diagnosis_ICD_10_C_M_Code',
    'Other_Diagnosis_Code_1_ICD_10_C_M',
    'Other_Diagnosis_Code_2_ICD_10_C_M',
    'Other_Diagnosis_Code_3_ICD_10_C_M',
    'Other_Diagnosis_Code_4_ICD_10_C_M',
    'Other_Diagnosis_Code_5_ICD_10_C_M'
]
```

```python
# -------------------------------
# 1. Prepare county race priors
# -------------------------------
df_county_race = df_final.groupby("COUNTYFIPS")[race_cols].mean().reset_index()
```

```python
df_county_race
```

```python
df= df_final.copy()
```

```python
# -------------------------------
# 2. Prepare Age/Gender likelihood modifiers
# -------------------------------
age_bins = [0,18,35,50,65,80,100]
df['age_group'] = pd.cut(df['Age'], bins=age_bins, right=False)
# Compute P(race|age_group, gender) from available data
df_age_gender = df.groupby(['age_group', 'Gender'])[race_cols].mean().reset_index()
df_age_gender = df_age_gender.rename(columns={col: col + "_agegender" for col in race_cols})
```

```python
# -------------------------------
# 3. Merge county priors and age/gender priors into main table
# -------------------------------
df_county_race_renamed = df_county_race.rename(columns={col: col + "_county" for col in race_cols})
df = df.merge(df_county_race_renamed, on="COUNTYFIPS", how="left")
df = df.merge(df_age_gender, on=['age_group','Gender'], how='left')
df = df.loc[:, ~df.columns.duplicated()]

# -------------------------------
# 4. Fill missing race columns using Bayesian update
# -------------------------------
missing_mask = df[race_cols].isna().any(axis=1)
if missing_mask.any():
    probs_matrix = df.loc[missing_mask, [r + '_county' for r in race_cols]] * \
                   df.loc[missing_mask, [r + '_agegender' for r in race_cols]].values

    # Normalize
    probs_matrix = probs_matrix.div(probs_matrix.sum(axis=1), axis=0)

    # Fill missing race columns
    max_race_idx = probs_matrix.idxmax(axis=1).str.replace('_county', '')
    for r in race_cols:
        df.loc[missing_mask, r] = (max_race_idx == r).astype(float)
```

```python
# -------------------------------
# 5. Cleanup
# -------------------------------
df.drop(columns=[r + '_county' for r in race_cols] + [r + '_agegender' for r in race_cols], inplace=True)
# -------------------------------
# 6. Check results
# -------------------------------
missing = df.isna().sum().sort_values(ascending=False)
for col, miss_count in missing.items():
    print(f"{col}: {miss_count}")

```

```python
# Count missing values per column and sort descending
missing = df.isna().sum().sort_values(ascending=False)

# Print all columns with their missing counts
for col, miss_count in missing.items():
    print(f"{col}: {miss_count}")
```

```python
all_cols = [hipps_col] + icd_cols

# Patient-level fill
for col in all_cols:
    missing_before = df[col].isna().sum()
    print(f"Missing {col} before patient-level fill: {missing_before}")
    
    df[col] = df.groupby('Beneficiary_ID')[col].transform(
        lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'UNKNOWN')
    )
    
    missing_after = df[col].isna().sum()
    print(f"Missing {col} after patient-level fill: {missing_after}\n")

# Population-level fill (age/gender group)
for col in all_cols:
    missing_before = df[col].isna().sum()
    print(f"Missing {col} before population-level fill: {missing_before}")
    
    df[col] = df.groupby(['age_group', 'Gender'], observed=True)[col].transform(
        lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'UNKNOWN')
    )
    
    missing_after = df[col].isna().sum()
    print(f"Missing {col} after population-level fill: {missing_after}\n")
```

```python
# Count missing values per column and sort descending
missing = df.isna().sum().sort_values(ascending=False)

# Print all columns with their missing counts
for col, miss_count in missing.items():
    print(f"{col}: {miss_count}")
```

```python
df = df_final.dropna().copy()

# ---------------------------
# 1. Extract Year and Month
# ---------------------------
df["Assessment_Effective_Date"] = pd.to_datetime(df["Assessment_Effective_Date"], errors="coerce")
df["Assessment_Effective_Year"]  = df["Assessment_Effective_Date"].dt.year
df["Assessment_Effective_Month"] = df["Assessment_Effective_Date"].dt.month

# Drop original date column (optional)
df = df.drop(columns=["Assessment_Effective_Date"])
```

```python
# Count missing values per column and sort descending
missing = df.isna().sum().sort_values(ascending=False)

# Print all columns with their missing counts
for col, miss_count in missing.items():
    print(f"{col}: {miss_count}")
```

```python
# ---------------------------
# 2. Identify dtypes
# ---------------------------
object_columns = df.select_dtypes(include=['object']).columns.tolist()

# Add the extra object columns properly
extra_object_cols = ['Facility_Internal_ID', 'Assessment_Effective_Year', 'Assessment_Effective_Month']

# Combine and remove Beneficiary_ID
columns = [col for col in object_columns if col != 'Beneficiary_ID'] + extra_object_cols

# Boolean columns
booleans = [
    'American_Indian_or_Alaska_Native', 'Asian', 'Black_or_African_American',
    'Hispanic_or_Latino', 'Native_Hawiian_or_Pacific_Islander', 'White', 'is_diabetes', 'is_heart_failure', 'is_hypertension', 'is_deceased'
]

# Force weight to be numeric
numeric_force = ['weight']

# Full categorical list EXCLUDING weight
categorical_cols = [c for c in (columns + booleans) if c not in numeric_force]

# Numeric columns = everything not categorical
numeric_cols = df.drop(columns=categorical_cols, errors='ignore').columns.tolist()
numeric_exclude = ['Beneficiary_ID', 'READMISSION'] + numeric_force
numeric_cols = [col for col in numeric_cols if col not in numeric_exclude]
print("Numeric columns:", numeric_cols, "\n")
print("Cat columns:", categorical_cols, "\n")
df_numeric = df[numeric_cols]

# ---------------------------
# 3. Split by cardinality
# ---------------------------
high_card_cols = [col for col in categorical_cols if df[col].nunique() > 55]
low_card_cols  = [col for col in categorical_cols if df[col].nunique() <= 55]

print("High-cardinality (Binary/Frequency Encoding):", high_card_cols, "\n")
print("Low-cardinality (One-Hot):", low_card_cols, "\n")
```

```python
# ---------------------------
# 4. Apply Binary Encoding
# ---------------------------
if len(high_card_cols) > 0:
    binary_encoder = ce.BinaryEncoder(cols=high_card_cols)
    df_binary = binary_encoder.fit_transform(df[high_card_cols])
else:
    df_binary = pd.DataFrame(index=df.index)

# ---------------------------
# 5. Apply One-Hot Encoding
# ---------------------------
if len(low_card_cols) > 0:
    df_onehot = pd.get_dummies(df[low_card_cols], drop_first=False)
else:
    df_onehot = pd.DataFrame(index=df.index)
```

```python
# ---------------------------
# 6. Combine Everything
# ---------------------------
df_encoded = pd.concat([df[['Beneficiary_ID', 'READMISSION']], df_numeric, df_binary, df_onehot], axis=1)

print("Final encoded shape:", df_encoded.shape)
df_encoded
```

```python
corr = df_encoded.corr(numeric_only=True).abs()
```

```python
threshold = 0.7

# Create an upper triangle boolean mask
mask = np.tril(np.ones(corr.shape), k=0).astype(bool)  # lower triangle including diagonal
upper = corr.where(~mask)  # keep only upper triangle

# Find features to drop
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

# Drop the features
df_reduced = df_encoded.drop(columns=to_drop).fillna(0).drop_duplicates()

print(f"Features dropped (correlation > {threshold}):")
print(to_drop)
print(f"Remaining features: {df_reduced.shape[1]}")
```
