```python
cols_to_drop = [
    'Beneficiary_ID',                      # Just an identifier
    'Agency_Medicare_Number',              # Which agency - not predictive
    'Primary_Diagnosis_ICD_10_C_M_Code',   # Already processed into flags
    'Other_Diagnosis_Code_1_ICD_10_C_M',
    'Other_Diagnosis_Code_2_ICD_10_C_M',
    'Other_Diagnosis_Code_3_ICD_10_C_M',
    'Other_Diagnosis_Code_4_ICD_10_C_M',
    'Other_Diagnosis_Code_5_ICD_10_C_M'
]
```

```python
# Check which columns exist before dropping
cols_to_drop = [col for col in cols_to_drop if col in df.columns]

df = df.drop(columns=cols_to_drop)

print(f"Dropped {len(cols_to_drop)} columns")
print(f"New shape: {df.shape}")
```

```python
# Convert to datetime
df['Assessment_Effective_Date'] = pd.to_datetime(df['Assessment_Effective_Date'])
```

```python
# Extract features
df['Assessment_Year'] = df['Assessment_Effective_Date'].dt.year
df['Assessment_Month'] = df['Assessment_Effective_Date'].dt.month
df['Assessment_DayOfWeek'] = df['Assessment_Effective_Date'].dt.dayofweek  # 0=Monday, 6=Sunday
```

```python
# Drop original date
df = df.drop(columns=['Assessment_Effective_Date'])
```

```python
print(f"\nNew shape: {df.shape}")
```

```python
df[['Assessment_Year', 'Assessment_Month', 'Assessment_DayOfWeek']].head()
```

```python
# One-Hot Encoding for low cardinality columns
onehot_cols = ['Gender', 'ByDiscipline', 'BMI_Category']

df = pd.get_dummies(df, columns=onehot_cols, drop_first=False)

print(f"\nAfter one-hot encoding, shape: {df.shape}")
```

```python
# Binary Encoding for high cardinality columns
import category_encoders as ce

binary_cols = ['COUNTY_NAME', 'Submitted_HIPPS_Code']

# Check which columns exist
binary_cols = [col for col in binary_cols if col in df.columns]

if len(binary_cols) > 0:
    print(f"Applying binary encoding to: {binary_cols}")
    
    # Create binary encoder
    binary_encoder = ce.BinaryEncoder(cols=binary_cols)
    
    # Fit and transform
    df = binary_encoder.fit_transform(df)
    
    print(f"\nAfter binary encoding, shape: {df.shape}")
    print(f"New columns created: {[col for col in df.columns if any(bc in col for bc in binary_cols)]}")
else:
    print("No columns to binary encode")
```

```python
# Checking data types after cleaning
print(df.dtypes.value_counts())

object_cols = df.select_dtypes(include=['object']).columns.tolist()

if len(object_cols) > 0:
    print(f"\Still have {len(object_cols)} object columns:")
    print(object_cols)
else:
    print("\nAll columns are numeric!")
```

```python
# Checking for missing values
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing_Count': missing.values,
    'Missing_Percent': missing_pct.values
})

missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

if len(missing_df) > 0:
    print(f"\nFound {len(missing_df)} columns with missing values:")
    print(missing_df.head(20))
else:
    print("\nNo missing values found!")
```

```python
# Find race columns
race_keywords = ['indian', 'asian', 'black', 'african', 
                 'hispanic', 'latino', 'hawaiian', 'pacific', 'white']

race_cols = [col for col in df.columns 
             if any(keyword in col.lower() for keyword in race_keywords)]

print(f"Found {len(race_cols)} race columns:\n")

# Create summary dataframe
summary = []
for col in race_cols:
    summary.append({
        'Column': col,
        'Missing_Count': df[col].isna().sum(),
        'Missing_Pct': f"{df[col].isna().sum()/len(df)*100:.1f}%",
        'Has_Data_Count': df[col].notna().sum(),
        'Unique_Values': df[col].nunique()
    })

summary_df = pd.DataFrame(summary)
print(summary_df.to_string(index=False))

# Check if missing values are same patients across all race columns
print("\n" + "="*60)
print("Are the SAME patients missing across all race columns?")
if len(race_cols) > 0:
    first_col_missing = df[race_cols[0]].isna()
    all_same = all(df[col].isna().equals(first_col_missing) for col in race_cols)
    print(f"{'YES same patients' if all_same else 'NO different patients'}")
```

```python
# Step 1: Create race_unknown flag
patient_race_cols = ['American_Indian_or_Alaska_Native', 'Asian', 
                     'Black_or_African_American', 'Hispanic_or_Latino',
                     'Native_Hawiian_or_Pacific_Islander', 'White']

df['race_unknown'] = df[patient_race_cols[0]].isna().astype(int) # Does 1 = "race data missing" and 0 = "have race data"

print(f"Created race_unknown flag:")
print(df['race_unknown'].value_counts())
```

```python
# Step 2: Fill missing values with 0 for patient race columns
for col in patient_race_cols:
    df[col] = df[col].fillna(0)
    
print("Imputed missing race values with 0")
print("\nCheck for NaN in race columns")
for col in patient_race_cols:
    print(f"{col}: {df[col].isna().sum()} missing")
```

```python
# Step 3: Drop columns that are missing much data
cols_to_drop = [
    'Unknown_Race_Ethnicity',  # 100% missing
    'DaysBetweenVisits',       # 99.93% missing
    'Weight_in_pounds',        # 62.72% missing (have BMI already)
    'Height_in_inches'         # 62.37% missing (have BMI already)
]

# Check which columns exist
cols_to_drop = [col for col in cols_to_drop if col in df.columns]

print(f"Dropping {len(cols_to_drop)} columns:")
for col in cols_to_drop:
    print(f"  - {col}")

df = df.drop(columns=cols_to_drop)

print(f"\nDropped:")
print(f"New shape: {df.shape}")
```

```python
# Check remaining missing values
missing = df.isnull().sum()
missing_df = pd.DataFrame({
    'Column': missing[missing > 0].index,
    'Missing_Count': missing[missing > 0].values,
    'Missing_Percent': (missing[missing > 0] / len(df) * 100).values
}).sort_values('Missing_Count', ascending=False)

print(f"Columns with missing values: {len(missing_df)}")
print("\n" + missing_df.to_string(index=False))
```

```python
"""
NOTE (2024-11-27):
- is_heart_failure and is_hypertension are 100% identical (29,605 patients)
- Correlation = 1.0, keeping both wastes resources and inflates feature importance
- DECISION: Dropping is_hypertension, keeping is_heart_failure
"""

df = df.drop(columns=['is_hypertension'])
print(f"Shape after dropping duplicate: {df.shape}")
print(f"Remaining features: {df.shape[1]}")
```

```python
df['Gender'] = df['Gender_Female'] 
df = df.drop(columns=['Gender_Male', 'Gender_Female'])

print(f"Shape after gender consolidation: {df.shape}")
print(f"Gender distribution:\n{df['Gender'].value_counts()}")

df['Gender'] = df['Gender'].astype(int)

print("Gender distribution (fixed):")
print(df['Gender'].value_counts())
print(f"\n1 = Female: {(df['Gender']==1).sum():,} ({(df['Gender']==1).sum()/len(df)*100:.1f}%)")
print(f"0 = Male: {(df['Gender']==0).sum():,} ({(df['Gender']==0).sum()/len(df)*100:.1f}%)")
```

```python
# Drop perfectly correlated features (r ≥ 0.999)
perfect_corr_drops = [
    'POP_COU',    # Keep POP_URB 
    'HOU_COU',    # Redundant with POP_URB
    'ACS_TOT_CIVIL_NONINST_POP_POV',  # Keep ACS_TOT_POP_POV
    'ACS_TOT_CIVIL_POP_POV',          # Keep ACS_TOT_POP_POV
    'ACS_PCT_HH_BROADBAND_ANY',       # Keep ACS_PCT_HH_INTERNET
    'ACS_PCT_HH_FOOD_STMP',           # Keep ACS_PCT_HH_PUB_ASSIST
]
df = df.drop(columns=[c for c in perfect_corr_drops if c in df.columns])
print(f"Dropped {len(perfect_corr_drops)} perfectly correlated features")
print(f"New shape: {df.shape}")
```

```python
# Create hipps_unknown flag : check any of the binary encoded HIPPS columns
hipps_cols = [col for col in df.columns if 'Submitted_HIPPS_Code_' in col]
if len(hipps_cols) > 0:
    # If any HIPPS binary column has NaN, mark as unknown
    df['hipps_unknown'] = df[hipps_cols[0]].isna().astype(int)
    
    # Fill NaN in all HIPPS binary columns with 0
    df[hipps_cols] = df[hipps_cols].fillna(0)
    
    print(f"Created hipps_unknown flag")
    print(f"Filled {len(hipps_cols)} HIPPS binary columns with 0")
else:
    print("No HIPPS columns found")
```

```python
# Check for any missing values
missing = df.isnull().sum()
total_missing = missing.sum()

print(f"Total missing values: {total_missing}")

if total_missing > 0:
    missing_df = pd.DataFrame({
        'Column': missing[missing > 0].index,
        'Missing_Count': missing[missing > 0].values
    })
    print("\nColumns with missing values:")
    print(missing_df.to_string(index=False))
else:
    print("\nNo missing values!")

print(f"\Dataset shape: {df.shape}")
print(f"Total rows: {df.shape[0]:,}")
print(f"Total columns: {df.shape[1]}")
```
